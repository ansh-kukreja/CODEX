```python
import json
```

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)

    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        # Cleaning Ratings data
        raw_rating = str(user["rating"]).strip().lower()
        
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        
        # Fixing inconsistent data
        raw_age = user.get("age")

        if(raw_age == None):
            user["age"] = None


        # De-Duplication of data
        if(user["name"].strip() in unique_users):
            continue

        unique_users.add(user["name"].strip())
        cleaned_data.append(user)
        
    return cleaned_data
```

```python
data = clean_data(data)
```

```python
def get_insights(data):
    # Average Ratings
    tot_rating = 0

    for user in data:
        tot_rating += float(user["rating"])

    print(f"Avg Ratings: {tot_rating/len(data)}")

    
    # Poor Ratings
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_ratings += 1

    print(f"Percentage of Poor Ratings: {poor_ratings / len(data) * 100}%")
    
```

```python
get_insights(data)
```

```python
def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        
        if(float(user["rating"]) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "OnePlus"

        recommendations.append(curr_recomm)

    return recommendations
```

```python
get_recommendations(data)
```
