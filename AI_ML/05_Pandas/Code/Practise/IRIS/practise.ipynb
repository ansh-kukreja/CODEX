```python
import pandas as pd
```

```python
data = pd.read_csv("IRIS.csv")

print(data.head(10))
print(type(data))
print(data.shape)
print(data.info())
print(data.describe())
```

```python
data[ (data["petal_length"] > 4.5) & (data["species"] == "Iris-virginica")]
```

```python
data.groupby("species").agg(
    Avg_Sepal_Length=("sepal_length", "mean"),
    Max_Petal_Width=("petal_width", "max"),
    StDev_Sepal_Width=("sepal_width", "std")
)
```

```python
data["petal_ratio"] = data["petal_length"] / data["petal_width"]

data.head(10)
```
