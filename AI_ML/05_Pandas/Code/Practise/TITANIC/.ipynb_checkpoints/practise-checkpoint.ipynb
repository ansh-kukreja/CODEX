```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic.csv")

df.columns

df[(df["Sex"] == "female") & (df["Fare"] > 30)]
df.query("Sex == 'female' & Fare > 30")
```

```python
df.groupby("Pclass").agg(
    Survival_Rate=("Survived", lambda x: x.mean() * 100),
    Avg_Fair=("Fare", "mean"),
    Avg_Age=("Age", "mean")
)
```
