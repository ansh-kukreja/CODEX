### Comparing speeds: Python Lists vs NumPy Arrays

```python
import numpy as np
import time
```

```python
size = 10_000_000

#python list
py_list = list(range(size))
start = time.time()
sq_list = [x**2 for x in py_list]
end = time.time()
print(f"python list time = {end-start} seconds")


#numpy arrays
np_array = np.array(py_list)
start = time.time()
sq_array = np_array ** 2
end = time.time()
print(f"numpy array time = {end-start} seconds")
```

#### As a Result: NumPy Arrays are way Faster than Python List
#### but NumPy Arrays can only store homogeneous data

### Comparing Memory consumption

```python
import sys

print("Python list size:", sys.getsizeof(py_list) * len(py_list))
print("NumPy array size:", np_array.nbytes)
```
